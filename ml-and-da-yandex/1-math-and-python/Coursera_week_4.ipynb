```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
```

## Экспоненциальное распределение

```python
# зададим случайную величину, соответсвующую распределению
rv = sts.expon(loc=0, scale=1)
# сгенерируем выборку 1000 значений
r = rv.rvs(size=1000)
```

```python
# построим гистаграмму по выборке объема 1000 

plt.hist(r, density=True, label='Exponential actual') # normed - Deprecated; use the density keyword argument instead.

# и теоретическую плотность распределения поверх нее
x = np.arange(0, 15, 0.001)
pdf=rv.pdf(x)
plt.plot(x, pdf, label='Exponential theor')

plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```

```python
# гистограммы распределений выборочных средних и приближающих их нормальных распределений при значении выборки - 10
mean = []
sample_size = 10
for i in range(1000):
    rv = sts.expon(loc=0, scale=1)
    r = rv.rvs(size=sample_size)
    mean.append(r.mean())
    

plt.hist(mean, density=True, label='Exponential actual')
x = np.arange(0, 5, 0.001)
#plt.plot(x, rv.pdf(x), label='Exponential theor')


# расчитаем теоритеческую плотность вероятности (pdf - probability density function) для нормального распределения
mu = 1 # theoretical expectation
sigma = (1./sample_size) ** 0.5 # theoretical sigma
norm_rv = sts.norm(mu,sigma)
pdf = norm_rv.pdf(x)

plt.plot(x,pdf,label='Norm theor')
plt.legend()
plt.ylabel('$f(x)$')
plt.xlabel('$x$')
plt.show()
```
