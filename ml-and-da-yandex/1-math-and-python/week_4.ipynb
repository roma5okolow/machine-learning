## 1. Построение выборки экспоненциального распределения

```python
import numpy as np
import pandas as pd
import scipy.stats as sts
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
expon_rv = sts.expon(loc = 0, scale = 1)
sample = expon_rv.rvs(1000)
plt.hist(sample, density = True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
```

```python
x = np.linspace(0, 7, 1000)
pdf = expon_rv.pdf(x)
plt.hist(sample, density = True, bins = 30)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.plot(x, pdf)
```

## 2. Распределение выборочного среднего

$$f(x) = \lambda \cdot \exp(- \lambda x)$$

$$\lambda = 1$$

$$E(X) = 1 / \lambda = 1$$

$$D(X) = 1 / \lambda^2 = 1$$

### 2.1.1 Построение гистограммы распределения выборочного среднего при n = 5 

```python
n1 = 5
sample_1 = np.random.choice(sample, n1)
print(sample_1,np.mean(sample_1) )
```

```python
sample_1_mean = []
for k in range(1000):
    sample_1 = np.random.choice(sample, n1)
    sample_1_mean.append(np.mean(sample_1))
```

```python
plt.hist(sample_1_mean, density = True, bins = 20)
plt.ylabel('pdf')
plt.xlabel('$mean 1$')
```

### 2.1.2 Построение нормального распределения:

```python
x = np.linspace(0, 3, 1000)
norm_1 = sts.norm(loc = 1, scale = np.sqrt(1/n1))
norm_1.rvs(10)
pdf = norm_1.pdf(x)
```

```python
plt.hist(sample_1_mean, density = True, bins = 20)
plt.plot(x, pdf)
plt.ylabel('pdf')
plt.xlabel('$mean 1$')
plt.show()
```
