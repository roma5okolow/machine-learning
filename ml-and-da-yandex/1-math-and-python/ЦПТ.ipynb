```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.stats as sts
import math
```

# Распределение $\chi^2$ (хи-квадрат) с $k$ степенями свободы, где $k=10$

```python
chi_1 = sts.chi2(10)  
```

## Генерирование выборки объемом 1000 значений

```python
sample = chi_1.rvs(1000)
x = np.linspace(1,50,100)
pdf = chi_1.pdf(x)
plt.hist(sample,density = True,label = 'Гистограмма выборки')
plt.plot(x,pdf,label = "Плотность вероятности")
plt.legend
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.show()
```

## Оценка распределение выборочного среднего вашей случайной величины при разных объёмах выборок

```python
for n in [5,10,50]:
    sample_values = []
    for i in range(1000):
        sample_values.append(chi_1.rvs(n).mean())
    plt.hist(sample_values, density = True, bins = 20, range = [6,12])
    norm = sts.norm(10, np.sqrt(2.*10./n))
    x = np.linspace(0,15,100)
    pdf2 = norm.pdf(x)
    plt.plot(x,pdf2)
    plt.show()
```

## Вывод:
### При увеличении размера выборки n, распределение средних значений начинает приближаться к 10 , а среднеквадратичное отклонение падает. Это значит,что точность апроксимации прямопропорциональнавеличине n (размер выборки)

### Результат не противоречит закону больших чисел,т.к. расчетное нормальное распределение визуально похоже на гистограмму
