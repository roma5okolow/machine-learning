```python
import numpy as np
import pandas as pd
from sklearn import cluster
```

```python
dataset = pd.read_csv("checkins.dat", sep = "|", header = 0, skipinitialspace = True)
dataset.dropna(inplace = True)
dataset.head(10)
```

```python
dataset.shape
```

```python
data = dataset[:100000]
data = data.loc[:, data.columns[3:5]]
data.head(10)
data.shape
```

```python
clustering = cluster.MeanShift(bandwidth = 0.1, n_jobs = 6).fit(data)
```

```python
centers = clustering.cluster_centers_
print(centers, centers.shape)
```

```python
clustering.labels_.shape
result = dict(zip(*np.unique(clustering.labels_, return_counts = True)))
print(result)
filtered_results_dict = {k : v for k, v in result.items() if v > 15}
print()
print(filtered_results_dict)
```

```python
# Для нанесения меток на карту, опционально
np.savetxt('test.txt', clustering.cluster_centers_, fmt = '%.4f')
```

```python
offices = pd.read_csv('offices.dat', header = 0, sep = ',')
offices.values
```

```python
def CalculateDistance (data, oficces):
    dists = np.zeros((data.shape[0], offices.shape[0]))
    for i in range(data.shape[0]):
        dist_to_one_point = np.zeros(offices.shape[0])
        for j in range(offices.shape[0]):
            dist_to_one_point[j] = np.linalg.norm(data[i] - offices.values[j])
        dists[i] = dist_to_one_point
    return dists
```

```python
# восстанавливаем точки по labels
filtered_data = np.zeros((len(filtered_results_dict), 2))
keys = list(filtered_results_dict.keys())
print(keys)
print("keys.shape = ", len(keys), "centers.shpae = ", centers.shape, "filtered_data.shape", filtered_data.shape)

for ind, key in enumerate(keys):
    filtered_data[ind, 0] = centers[key, 0]
    filtered_data[ind, 1] = centers[key, 1]
print(filtered_data)
```

```python
print(data, '\n', offices)
print(data.values[5] - offices.values[5], type(np.linalg.norm(data.values[5] - offices.values[5])))
dists = CalculateDistance(filtered_data, offices)
```
